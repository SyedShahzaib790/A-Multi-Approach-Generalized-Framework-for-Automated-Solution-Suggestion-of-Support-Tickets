```python
bow  = pd.read_pickle('../data/BoW_results.pkl')
bow_tfidf  = pd.read_pickle('../data/Bow-tdidf_results.pkl')
word2vec_cbow = pd.read_pickle('../data/word2vec_cbow_results.pkl')
word2vec_skipgram = pd.read_pickle('../data/word2vec_skipgram_results.pkl')
fasttext_cbow  = pd.read_pickle('../data/fasttext_cbow_results.pkl')
fasttext_skipgram  = pd.read_pickle('../data/fasttext_skipgram_results.pkl')
fasttext_pretrained  = pd.read_pickle('../data/fasttext_pretrained_results.pkl')
gensen  = pd.read_pickle('../data/gensen_results.pkl')
infersent  = pd.read_pickle('../data/infersent_results.pkl')
lstm_single  = pd.read_pickle('../data/lstm_results.pkl')
lstm_multi = pd.read_pickle('../data/multi-lst_results.pkl')
bert  = pd.read_pickle('../data/bert_results.pkl')
```

```python
def plot_analysis(model_name, model):
#     model_name = 'CBOW'
    ax = model[model.columns[1:]].plot(xticks=model.index, marker='o')
    ax.set_title('{} Approach'.format(model_name), weight='bold', size=11)
    ax.set_xlabel("Number of Tickets Considered (K)", labelpad=10, size=10)
    ax.set_ylabel("Evaluation Score", labelpad=10, size=10)
    ax.grid(True,linewidth=0.4)
    # plt.legend(model.columns[1:],loc='best', bbox_to_anchor=(1.0, 1.0),fontsize=12)
    plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.20),fancybox=True, shadow=True, ncol=model.columns.shape[0]-1)

    # plt.savefig('../data/cbow.jpg', format='jpg', quality=100, dpi=300, pad_inches=5)
    plt.show()
```

```python
models = {
    'Bag of Words (BoW)':bow,
    "Bag of Words (BoW) with TF-IDF":bow_tfidf,
    "Word2Vec (Continous Bag of Words - CBOW)":word2vec_cbow,
    "Word2Vec (Skip-gram)":word2vec_skipgram,
    "FastText (CBOW)":fasttext_cbow,
    "FastText (Skip-gram)":fasttext_skipgram,
    "FastText Pretrained (Embeddings)":fasttext_pretrained,
    "Gensen Pretrained (Embeddings)":gensen,
    "InferSent Pretrained (Embeddings)":infersent,
    "BERT Pretrained (Embeddings)":bert
}
```
